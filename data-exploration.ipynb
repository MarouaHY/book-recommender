```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1 ,0)
books["age_of_book"]= 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
                      cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python


book_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]

```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words.loc[:, 'title_and_subtitle'] = (
    np.where(book_missing_25_words["subtitle"].isna(), 
             book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words.loc[:, 'tagged_description'] = (
    book_missing_25_words[["isbn13", "description"]]
    .astype(str)
    .agg(" ".join, axis=1)
)
```

This operation creates a new column tagged_description that combines each book's ISBN13 (Unique book identifier: Every distinct book edition has a unique ISBN-13)  
and description into a single string, separated by a space:

    Selects two columns: Takes the isbn13 and description columns

    Converts to strings: Ensures both columns are treated as text with .astype(str)

    Joins them: Combines them with a space separator using .agg(" ".join, axis=1)

    Stores result: Creates a new column tagged_description with the combined strings
    

```python
(
    book_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index = False)
)
```
